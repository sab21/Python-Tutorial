#### Working with large amounts of data

```python
import pandas as pd
df=pd.read_excel('pokemon1.xlsx')
new_df=pd.DataFrame(columns=df.columns);new_df
```

```python
for df in pd.read_csv('pokemon_data.csv',chunksize=100):
    print('Chunk DF')
    print(df.head())
```

```python
new_df=pd.DataFrame(columns=df.columns)

for df in pd.read_csv('pokemon_data.csv',chunksize=100):
    results=df.groupby(['Type 1']).mean()
    
    new_df=pd.concat([new_df, results])
    
new_df.tail()   
```
